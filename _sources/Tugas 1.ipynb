# 2. Membuat statistik dari data setiap atribut yang bertype numerik

Dalam tugas ini, kita akan melakukan percobaan perhitungan statistic pada data panjang kelopak bunga (sepal length). Cara memanggil data yang ingin dihitung yakni dataset['kelompok_data']

```python
panjang_kelopak = dataset['sepal_length']
```

# 2.1 Median

Median memiliki pengertian sebagai nilai data tengah. Penyajian data yang dimaksud lebih kepada hasil dari penelitian. Biasanya data tersebut disajikan dalam bentuk tabel, diagram ataupun daftar dan kerap disebut sebagai statistik 

Rumus Median :

Rumus Median dengan jumlah data ganjil

$$ Me = \frac{X(n+1)}{2} $$

Rumus Median dengan jumlah data genap:

$$ Me = \frac{X\frac{n}{2}+X(\frac{n}{2}+1)}{2} $$


```python
median = panjang_kelopak.median()
print("Median = ", median)
```

# 2.2 Mean

Mean adalah salah satu ukuran gejala pusat. Mean dapat dikatakan sebagai wakil kumpulan data. Menentukan mean dapat dilakukan dengan cara menjumlahkan seluruh nulai data, kemudian membaginya dengan banyaknya data.





Rumus Mean :


$$ mean = \frac{x1+x2+x3...xn}{n} $$

```python
mean = panjang_kelopak.mean()
print("Mean =", mean)
```

# 2.3 Modus

Modus adalah data yang paling sering muncul. Modus merupakan ukuran pemusatan untuk menyatakan fenomena yang paling banyak terjadi. Sekumpulan data yang diperoleh, memungkinkan untuk memiliki nilai modus yang tidak tunggal atau mungkin juga tidak memilikinya.

```python
modus = panjang_kelopak.mode()
print("Modus = ", modus)
```

# 2.4 Kuartil

Kuartil adalah nilai yang membagi data yang berurutan menjadi empat bagian yang sama banyak. Karena data terbagi menjadi empat bagian yang sama, artinya terdapat tiga nilai kuartil, yaitu kuartil bawah(Q1), kuartil tengah (Q2), dan kuartil atas (Q3).

```python
kuartil = panjang_kelopak.quantile()
print("Kuartil = ", kuartil)
```

# 3. Mendeskripsikan kemiringan / skwened dari data setiap variabel

Kemiringan atau kecondongan(skewness) adalah tingkat ketidaksimetrisan atau kejauhan simetri dari sebuah distribusi. Skewness diartikan sebagai kemiringan distribusi data

```python
skewn = panjang_kelopak.skew()
print("Skewn = ", skewn)
```
